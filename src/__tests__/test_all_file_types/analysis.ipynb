# Homework 3 — Binary Search Tree Analysis

In this notebook we'll implement a BST and visualize its properties.

```python
class TreeNode:
    def __init__(self, key):
        self.key = key
        self.left = None
        self.right = None

class BST:
    def __init__(self):
        self.root = None

    def insert(self, key):
        if not self.root:
            self.root = TreeNode(key)
        else:
            self._insert(self.root, key)

    def _insert(self, node, key):
        if key < node.key:
            if node.left is None:
                node.left = TreeNode(key)
            else:
                self._insert(node.left, key)
        else:
            if node.right is None:
                node.right = TreeNode(key)
            else:
                self._insert(node.right, key)

    def inorder(self):
        result = []
        self._inorder(self.root, result)
        return result

    def _inorder(self, node, result):
        if node:
            self._inorder(node.left, result)
            result.append(node.key)
            self._inorder(node.right, result)

    def height(self):
        return self._height(self.root)

    def _height(self, node):
        if not node:
            return 0
        return 1 + max(self._height(node.left), self._height(node.right))
```

```python
tree = BST()
values = [50, 30, 70, 20, 40, 60, 80]
for v in values:
    tree.insert(v)

print('Sorted:', tree.inorder())
print('Height:', tree.height())
```

## Performance Analysis

Let's measure how tree height grows with random insertions vs sorted insertions.

```python
import random

for n in [100, 500, 1000]:
    # Random insertion
    t1 = BST()
    shuffled = list(range(n))
    random.shuffle(shuffled)
    for v in shuffled:
        t1.insert(v)
    
    # Sorted insertion (worst case)
    t2 = BST()
    for v in range(n):
        t2.insert(v)
    
    print(f'n={n}:  random height={t1.height()}, sorted height={t2.height()}')
```

### Observation

Random insertions yield $O(\log n)$ height, while sorted insertions degrade to $O(n)$.
This motivates self-balancing trees like **AVL** and **Red-Black** trees.
