```python
#include <cassert>
#include <fstream>
#include <numeric>
#include <sstream>
#include <string>
#include <vector>

std::vector<int> parse_numbers(const std::string& csv_line) {
    std::vector<int> values;
    std::stringstream ss(csv_line);
    std::string token;
    while (std::getline(ss, token, ',')) values.push_back(std::stoi(token));
    return values;
}
```

```python
std::ifstream input("../existing_data.txt");
assert(input.good());
std::string line;
std::getline(input, line);
auto values = parse_numbers(line);
int sum = std::accumulate(values.begin(), values.end(), 0);
std::ofstream out("result_summary.txt"); out << "sum=" << sum << std::endl;
std::cout << "=== C++ NOTEBOOK RENDER TEST ===" << std::endl;
std::cout << "unicode: café λ 🚀" << std::endl;
std::cout << "idx | value" << std::endl;
for (size_t i = 0; i < values.size(); ++i) std::cout << i << " | " << values[i] << std::endl;
assert(sum == 50);
```
