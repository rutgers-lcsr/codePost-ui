# R Complete Notebook
Performs file I/O, plotting, assertions, and render-friendly console output.

```python
parse_numbers <- function(csv_line) {
  as.numeric(strsplit(csv_line, ',')[[1]])
}
```

```python
csv_line <- trimws(readLines('../existing_data.txt', warn = FALSE)[1])
values <- parse_numbers(csv_line)
writeLines(sprintf('sum=%d', sum(values)), 'result_summary.txt')
plot(values, type='b', col='steelblue', pch=19, main='R Notebook Render Plot')
cat('=== R NOTEBOOK RENDER TEST ===\n')
cat('unicode: café λ 🚀\n')
cat('idx | value\n')
for (i in seq_along(values)) cat(sprintf('%d | %d\n', i - 1, values[i]))
stopifnot(sum(values) == 50)
```
