Performs file I/O, assertions, and render-oriented output for UI checks.

```python
def parse_numbers(csv_line)
  csv_line.split(',').map { |part| Integer(part.strip) }
end
```

```python
csv_line = File.read('../existing_data.txt').strip
values = parse_numbers(csv_line)
sum = values.sum
File.write('result_summary.txt', "sum=#{sum}\n")
puts '=== RUBY NOTEBOOK RENDER TEST ==='
puts 'unicode: café λ 🚀'
puts 'idx | value'
values.each_with_index { |value, i| puts "#{i} | #{value}" }
raise 'Expected sum=50' unless sum == 50
```
