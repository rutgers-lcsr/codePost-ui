# Node.js Complete Notebook
Runs file I/O, assertions, and render-heavy console output.

```python
const fs = require('node:fs');

function parseNumbers(csvLine) {
  return csvLine.split(',').map((item) => Number(item.trim()));
}
```

```python
const csv = fs.readFileSync('../existing_data.txt', 'utf8').trim();
const values = parseNumbers(csv);
const sum = values.reduce((acc, value) => acc + value, 0);
fs.writeFileSync('result_summary.txt', `sum=${sum}\n`);
console.log('=== NODE NOTEBOOK RENDER TEST ===');
console.log('unicode: café λ 🚀');
console.log('idx | value');
values.forEach((value, index) => console.log(`${index} | ${value}`));
console.error('[stderr][node-notebook] Render check stderr line');
if (sum !== 50) throw new Error('Expected sum=50');
```
