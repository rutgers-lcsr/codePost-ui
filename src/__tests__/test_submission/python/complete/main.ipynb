# Python Complete Notebook
Performs file I/O, plotting, assertions, and render-oriented console output.

```python
import json
from pathlib import Path
import matplotlib.pyplot as plt

def parse_numbers(csv_line: str) -> list[int]:
    return [int(part.strip()) for part in csv_line.split(',')]
```

```python
csv_line = Path('../existing_data.txt').read_text(encoding='utf-8').strip()
values = parse_numbers(csv_line)
summary = {'count': len(values), 'sum': sum(values), 'max': max(values)}
Path('result_summary.json').write_text(json.dumps(summary, indent=2), encoding='utf-8')
plt.plot(values, marker='o')
plt.title('Python Notebook Render Plot')
plt.show()
print('=== PYTHON NOTEBOOK RENDER TEST ===')
print('unicode: café λ 🚀')
print('idx | value')
for i, value in enumerate(values):
    print(f'{i:>3} | {value}')
assert summary['sum'] == 50
```
