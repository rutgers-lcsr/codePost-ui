```python
import java.nio.file.Files;
import java.nio.file.Path;
import java.util.Arrays;
import java.util.List;
import java.util.stream.Collectors;

List<Integer> parseNumbers(String csvLine) {
  return Arrays.stream(csvLine.split(",")).map(String::trim).map(Integer::parseInt).collect(Collectors.toList());
}
```

```python
String csv = Files.readString(Path.of("../existing_data.txt")).trim();
List<Integer> values = parseNumbers(csv);
int sum = values.stream().mapToInt(Integer::intValue).sum();
Files.writeString(Path.of("result_summary.txt"), "sum=" + sum + System.lineSeparator());
System.out.println("=== JAVA NOTEBOOK RENDER TEST ===");
System.out.println("unicode: café λ 🚀");
System.out.println("idx | value");
for (int i = 0; i < values.size(); i++) System.out.println(i + " | " + values.get(i));
assert sum == 50;
```
