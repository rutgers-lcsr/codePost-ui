# PHP Complete Notebook
Runs file I/O, assertion checks, and render-friendly console output.

```python
<?php
function parse_numbers(string $csv_line): array {
  return array_map(static fn(string $item): int => (int) trim($item), explode(',', $csv_line));
}
```

```python
$csv = trim((string) file_get_contents('../existing_data.txt'));
$values = parse_numbers($csv);
$sum = array_sum($values);
file_put_contents('result_summary.txt', 'sum=' . $sum . PHP_EOL);
echo '=== PHP NOTEBOOK RENDER TEST ===' . PHP_EOL;
echo 'unicode: café λ 🚀' . PHP_EOL;
echo 'idx | value' . PHP_EOL;
foreach ($values as $idx => $value) { echo $idx . ' | ' . $value . PHP_EOL; }
fwrite(STDERR, '[stderr][php-notebook] Render check stderr line' . PHP_EOL);
assert($sum === 50, 'Expected sum=50');
```
